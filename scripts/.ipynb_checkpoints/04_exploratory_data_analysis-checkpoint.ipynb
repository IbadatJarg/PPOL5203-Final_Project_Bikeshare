```python
#Setup
import numpy as np
import pandas as pd
import json
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
```

```python
#read files
df = pd.read_csv("../data/cleaned_data/final_data.csv")

df = df[df['opened_station'] == 2013]
```

#### ADDT_sum

```python
#Independent variable
X = df[['opened_station']] 

# Dependent variables
Y = df[['change_AADT_mean']] 
Y = Y.dropna()

#Aligning our models
X, Y = X.align(Y, join='inner', axis=0)

# Fit the model
model_AADT = LinearRegression()
model_AADT.fit(X, Y)
```

```python
# Create a table of results
coefficients = pd.DataFrame(
    model_AADT.coef_,  # Coefficients for each dependent variable
    columns=['Coefficient'],  # Column name
    index=Y.columns  # Dependent variable names
)
coefficients['Intercept'] = model_AADT.intercept_  # Add intercepts for each dependent variable

# Display as a table
print(coefficients)
```

```python
# Add a constant for the intercept
X = sm.add_constant(X)

# Fit separate models for each dependent variable
results = {}
for col in Y.columns:
    model_AADT = sm.OLS(Y[col], X).fit()
    results[col] = model_AADT

# Extract coefficients and standard errors into a table
output = {}
for col, model_AADT in results.items():
    params = model_AADT.params  # Coefficients
    std_err = model_AADT.bse    # Standard errors
    output[col] = pd.DataFrame({
        "Coefficient": params,
        "Standard Error": std_err,
        "p-value": model_AADT.pvalues
    })

# Display results for each dependent variable
for col, table in output.items():
    print(f"\nResults for {col}:\n")
    print(table)
```
